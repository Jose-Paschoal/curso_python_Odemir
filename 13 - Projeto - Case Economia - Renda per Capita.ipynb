```python
## importando as lib

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## lendo a base

base = pd.read_excel('Dados_Pib.xlsx')

base.head()
```

```python
base.groupby(by=['Territorialidades', 'Ano']).mean()
```

```python
## sistema de grids

# cor de fundo

fundo = '#f5f5f5'

# sistema de grids

grid = sns.FacetGrid(base, col='Territorialidades', hue='Territorialidades', col_wrap=4)

# adicionar grafico de linhas em cada grafico

grid = grid.map(plt.plot, 'Ano', 'PIB per capita');

# adicionar uma sombra e mudança no nome dos gráficos

grid = grid.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

# filtrar o título

grid = grid.set_titles('{col_name}')

# adicionar subtitulo

grid = grid.fig.suptitle('Evolução da Renda per capita por Estado', fontsize=15)

# ajuste

plt.subplots_adjust(top = 0.92)
```
