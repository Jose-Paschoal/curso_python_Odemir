```python
provas = ['math score', 'reading score', 'writing score']
```

```python
## ordenar o dataset

df = df.sort_values(by = provas, ascending = False)\
.reset_index(drop = True)
```

```python
df
```

```python
## coluna com a mÃ©dia das provas

df['mean'] = df[provas].mean(axis=1)
```

```python
df.head()
```

```python
## fazendo uma consulta no dataset

df.query('(gender == "male") & ("test preparation course" == "none") & ("math score" >= 70)')
```

```python
df[(df.gender == 'male') & (df['test preparation course'] == 'none') & (df['math score'] >= 70)]
```

```python
## agrupamento -  agrupa os dados por gÃªnero e obtÃ©m estatÃ­sticas descritivas

df.groupby(by = 'gender')[provas].agg([np.mean, np.median]).T
```

## VisualizaÃ§Ã£o de Dados

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### GrÃ¡fico de Linhas

```python
np.random.seed(7)

## vetor com nÃºmeros inteiros aleatÃ³rios entre 1, 1500 com 10 amostras

y = np.random.randint(low = 1, high= 1500, size = 10)
```

```python
y
```

```python
plt.plot(y)
```

```python
## refinando o grÃ¡fico

# insere a primeira linha do plot

plt.plot(y, color = '#749187', marker = 'o', ms = 5, mec = 'k', markerfacecolor = 'w', ls = '-.')

# insere a segunda linha

plt.plot(y*2, marker = '+', color = 'k', ms = 5)

# rÃ³tulos 

plt.xlabel('Eixo x', color = 'red', size = 12)
plt.ylabel('Eixo y')
plt.title('TÃ­tulo', loc = 'left')

# gridlines

plt.grid(axis = 'y', color = 'gray', linestyle = '--', linewidth = 1, alpha = 0.8)

plt.show()
```

### Subplots

```python
np.random.seed(6)
x = np.arange(1, 11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)

plt.figure(figsize = (15,5))
plt.suptitle('Figura', fontsize = 15)

plt.subplot(1,3,1)
plt.plot(x, y1, color = 'k')
plt.title('Subplot 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1,3,2)
plt.plot(x, y2, color = 'r')
plt.title('Subplot 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1,3,3)
plt.plot(x, y3, color = 'g')
plt.title('Subplot 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.tight_layout(pad = 4)
plt.show()
```

### Histograma e Boxplot

```python
## scatter plot

x = np.random.normal(10, 0.5, 100)
y = np.random.uniform(0, 20, 100)
```

```python
fig = plt.figure()
ax = plt.axes()
ax.scatter(x, y, marker = 'o', color = 'red', label = 'data 1', alpha = 0.5)
ax.scatter(x*0.5, y*0.5, marker ='v', color = 'black', label = 'data 2', alpha = 0.7)
ax.legend();
```

## Utilizandos Dados EconÃ´micos

```python
import plotly.express as px
```

```python
px.data.gapminder()
```

```python
## gravando os dados em um df e utilizando uma quebra de filtro

df = px.data.gapminder().query('country == "Brazil"').set_index('year')
```

```python
df.head()
```

```python
## EvoluÃ§Ã£o do PIB per capita no tempo

plt.plot(df.index, df['gdpPercap'])
plt.title('PIB per capita do Brasil')
plt.ylabel('PIB per capita')
plt.xlabel('Tempo');
```

```python
## Vendo relaÃ§Ã£o entre expectativa de vida e renda per capita

title = 'RelaÃ§Ã£o entre expectativa de vida e renda per capita no Brasil'

plt.figure(figsize = (12,4))
plt.scatter(df['lifeExp'], df['gdpPercap'], cmap= 'viridis')
plt.xlabel('Expectativa de vida')
plt.ylabel('Renda per capita')
plt.title(title, loc = 'left')
plt.show()
```

```python
plt.bar(x = df.index, 
       height = df['pop'],
       color = 'red')

plt.title('PopulaÃ§Ã£o brasileira');
```

## Checando Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
## Definindo um tema

sns.set_theme(style = 'darkgrid')
```

```python
## Dados do prÃ³prio Seaborn

dados = sns.load_dataset('tips')
```

```python
dados
```

```python
# Renomeando as colunas

dados.rename(columns ={'total_bill' : 'Total Conta', 'tip' : 'Gorjeta', 'sex' : 'Sexo', 'smoker' : 'Fumante', 'day' : 'Dia Semana', 'time' : 'PerÃ­odo', 'size' : 'Pessoas na Mesa'}, inplace = True)
```

```python
dados.head()
```

```python
# GrÃ¡fico Relplot

sns.relplot(x='Total Conta', y='Gorjeta', data = dados);
```

```python
## Passando outro parÃ¢metor como classe no grÃ¡fico

sns.relplot( x ='Total Conta', y = 'Gorjeta', data = dados, hue = 'Sexo');
```

```python
## Mudando formato do grÃ¡fico

sns.relplot( x ='Total Conta', y = 'Gorjeta', data = dados, kind = 'line');
```

```python
## Inserindo parÃ¢metro dentro do grÃ¡fico no formato de linhas

sns.relplot( x ='Total Conta', y = 'Gorjeta', data = dados, kind = 'line', hue = 'Fumante');
```

### GrÃ¡fico Histplot

```python
sns.histplot(data = dados, x='Total Conta');
```

```python
sns.histplot(data = dados, x='Total Conta', hue = 'Fumante');
```

### GrÃ¡fico Barplot

```python
sns.barplot(data = dados, x = 'Sexo', y = 'Total Conta', hue = 'Fumante');
```

### GrÃ¡fico Pairplot

```python
sns.pairplot(dados);
```

```python
sns.pairplot(dados, hue = 'Sexo');
```

```python
sns.pairplot(dados, kind = 'kde');
```

### GrÃ¡fico Boxplot

```python
sns.boxplot( data = dados, x = 'Dia Semana', y = 'Total Conta');
```
