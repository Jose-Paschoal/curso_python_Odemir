# Empresas de Energia 1

```python
## importando as lib

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados = pd.read_excel('Dados_Empresas_Energia.xlsx')

dados.head()
```

```python
## setar o index

dados.set_index('Data', inplace=True)

dados.head()
```

```python
## grafico

# estilo
plt.style.use('seaborn-darkgrid')

# tamanho
plt.figure(figsize=(16,7))

#titulo
plt.title('Análise ações de empresas de Energia', fontsize=18, fontweight=600)

# plot da petrobras
plt.plot(dados.index, dados['Petrobras'], linewidth=4, alpha=0.7)
# texto da petrobras
plt.text(dados.index[-1], dados['Petrobras'].tail(1), 'Petrobras', size='large', horizontalalignment='left')

# plot de todas as colunas
for coluna in dados.columns[1:]:
    # plot das outras colunas
    plt.plot(dados.index, dados[coluna],color='gray', linewidth=2, alpha=0.7)
    # texto das outras colunas
    plt.text(dados.index[-1], dados[coluna].tail(1), coluna, color='gray', size='large', horizontalalignment='left')


#labels
plt.xlabel('Período')
plt.ylabel('Preço de fechamento (R$)');
```
